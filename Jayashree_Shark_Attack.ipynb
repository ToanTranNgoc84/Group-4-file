```python
import pandas as pd
import numpy as np

# Load Shark Attack Data Set
url = "https://www.sharkattackfile.net/spreadsheets/GSAF5.xls"

shark_df=pd.read_excel(url)
display(shark_df)
```

**Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.**

```python
print(red_color+f"Before cleaning{reset_color}\n{shark_df.columns}")
```

**First Procedure**

```python
#Cleaning the column name:
#1. Strip unwanted spaces
#2. Replace space with underscore.
shark_df.rename(columns ={col :col.strip().lower().replace(' ','_') for col in shark_df.columns}, inplace = True)
print(red_color+f"After cleaning{reset_color}\n{shark_df.columns}")
```

**Second Procedure**

```python
#Cleaning sex column values
unique_sex_values = shark_df['sex'].unique()
print(red_color+f"Unique sex values before clenaing:{reset_color} {shark_df['sex'].unique()}")
#Convert values to upper case for better understanding
shark_df['sex'] = shark_df['sex'].str.upper().str.strip()
# Replace non-standard values with NaN
shark_df['sex'].replace({
    'LLI': np.nan,   # Replace 'LLI' with NaN
    'M X 2': 'M',    # Replace 'M X 2' with 'M'
    'N': np.nan,     # Replace 'N' with NaN
    '.': np.nan      # Replace '.' with NaN
}, inplace=True)
shark_df['sex'].fillna('Unknown', inplace=True)
# Print all unique values in sex column
print(red_color+f"Unique sex values after clenaing:{reset_color} {shark_df['sex'].unique()}")
```

```python
# Fill missing values in 'activity' column with 'Unknown'
print(red_color+f"Unique activity values before clenaing:\n"+reset_color)
display(shark_df['activity'].unique())
shark_df['activity'].fillna('Unknown', inplace=True)
shark_df['activity'] = shark_df['activity'].apply(lambda x: x.strip().replace('  ',' '))
print(red_color+f"\nUnique activity values after clenaing:\n"+reset_color)
display(shark_df['activity'].unique())
```

```python
# filling missing values in 'location' columns with 'Unknown'
print(red_color+f"Unique location values before clenaing:"+reset_color)
print(shark_df['location'].unique())
shark_df['location'].fillna('Unknown', inplace=True)
shark_df['location'] = shark_df['location'].apply(lambda x: x.strip().replace('  ',' '))
print(red_color+f"\nUnique location values after clenaing:"+reset_color)
print(shark_df['location'].unique())
```

```python
# filling missing values in 'species' columns with 'Unknown'
print(red_color+f"Unique species values before clenaing:"+reset_color)
print(shark_df['species'].unique())
shark_df['species'].fillna('Unknown', inplace=True)
shark_df['species'] = shark_df['species'].apply(lambda x: x.strip().replace('  ',' '))
print(red_color+f"\nUnique species values after clenaing:"+reset_color)
print(shark_df['species'].unique())
```

**Third Procedure**

```python
#change date format from date column
shark_df["date"] = pd.to_datetime(shark_df["date"], format="%d %b-%Y", errors='coerce')
# Format 'date' column as "15/03/24"
shark_df["date"] = shark_df["date"].dt.strftime("%d/%m/%y")
display(shark_df)
```

**Fourth Procedure**

```python
#change year data type
shark_df["year"].fillna(0, inplace = True)
shark_df["year"] = shark_df["year"].astype(int)
shark_df["year"].unique()
```

**Fifth Procedure**

```python
#Few columns don't seem to be required for our analysis. So, we are dropping them
shark_df.drop(columns=['unnamed:_11', 'unnamed:_21', 'unnamed:_22','case_number','case_number.1','pdf','href','href_formula','original_order'], inplace=True)
```

**Sixth Procedure**

```python
print(red_color+f"Before applying duplicate :{reset_color} {shark_df.duplicated().sum()}")
# Removing Duplicates
shark_df.drop_duplicates(inplace=True)
print(red_color+f"After applying duplicate :{reset_color} {shark_df.duplicated().sum()}")
```

**Seventh Procedure**

```python
# make column "country" consistent first letter upper, everything else lower except USA
print(red_color+f"BEFORE{reset_color}\n{shark_df['country']}")
shark_df["country"] = shark_df["country"].apply(lambda x: x.strip().title() if isinstance(x, str) else x)
# Becuase when we use title() the country names are converted to first letter upper case and rest lower case
shark_df["country"] = shark_df["country"].replace({'Usa':"USA"})
print(red_color+f"AFTER{reset_color}\n{shark_df['country']}")
```

**GROUPING**

```python
attacks_by_country = shark_df.groupby('country').size().sort_values(ascending=False)
print(red_color+f"Attacks by country {reset_color}\n{attacks_by_country}")
attacks_by_year = shark_df.groupby('year').size()
print(red_color+"Attacks by year"+reset_color)
print(attacks_by_year)
print()
attacks_by_sex = shark_df.groupby('sex').size().sort_values(ascending=False)
print(red_color+"Attacks by sex"+reset_color)
print(attacks_by_sex)
print()
attacks_by_species = shark_df.groupby('species').size().sort_values(ascending=False)
print(red_color+"Attacks by species"+reset_color)
print(attacks_by_species)
print()
attacks_by_activity = shark_df.groupby('activity').size().sort_values(ascending=False)
print(red_color+"Attacks by activity"+reset_color)
print(attacks_by_activity)
print()
attacks_by_year_injury = shark_df.groupby(['year'])[['injury']].count()
print(red_color+"Attacks by year & injury"+reset_color)
print(attacks_by_year_injury)
print()
```
